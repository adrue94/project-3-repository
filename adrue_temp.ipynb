```python
# Load CSV file and turn it into PD DataFrames, dropping and renaming columns
birth_DFS = pd.read_csv("data/8_states_2016_2021.csv")
birth_DFS
```

```python
birth_DFS.rename(columns={"Education_Level_of_Mother": "EduLevel", "Education_Level_Code": "EduLevelCode", "Number_of_Births": "Births", "Average_Age_of_Mother_(years)": "AvgMotherAge", "Average_Birth_Weight_(g)": "AvgBirthWeight"}, inplace=True)
birth_DFS.drop(columns = ["mean_births"], inplace=True)
birth_DFS
```

```python
merged_genders_DFS = pd.read_csv("data/merged_genders_2016_2021.csv")
merged_genders_DFS.rename(columns = {"Education_Level_of_Mother": "EduLevel", "Education_Level_Code": "EduLevelCode", "Number_of_Births": "Births"}, inplace=True)
merged_genders_DFS
```

```python
merged_years_DFS = pd.read_csv("data/merged_years.csv")
merged_years_DFS.rename(columns = {"Education_Level_of_Mother": "EduLevel", "Education_Level_Code": "EduLevelCode", "Number_of_Births": "Births"}, inplace=True)
merged_years_DFS
```
